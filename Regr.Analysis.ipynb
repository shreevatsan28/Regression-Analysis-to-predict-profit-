```python
#PROJECT STATEMENT:Perform regression analysis to predict the profit of each of the 50 startups mentioned in the dataset by taking all the input parameters(multi linear regression).
```

```python
#Creating a dataframe with the Dataset given
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
df
```

```python
#(rows,columns)
df.shape
```

```python
#total no. of records
df.size
```

```python
df.info()
```

```python
#to check wether the given dataset has null values or not
df.isnull().sum()
```

```python
#converting the categorical data into dummies or numerical data
df_dc = pd.get_dummies(df.State)

df_dc
```

```python
merged= pd.concat([df,df_dc],axis='columns')
merged
```

```python
final=merged.drop(['State','New York','California','Florida'],axis='columns')
final
```

```python
#Data Visualization of the input parameters
import seaborn as sns
sns.distplot(final['R&D Spend'],color='red')
```

```python
sns.distplot(final['Administration'],color='blue')
```

```python
sns.distplot(final['Marketing Spend'],color='green')
```

```python
sns.distplot(final['Profit'],color='cyan')
```

```python
# taking the input and output varaibles as x and y
x = final.iloc[:, 0:3] #columns other than profits are taken as the feature/inputs
y = final.iloc[:, -1] #profit is the output
```

```python
x
```

```python
y
```

```python
#now we split the data into test and train values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
#now importing the LinearRegression model and predicting the profit using multi linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
#fitting the data into the model
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
y_pred
```

```python
#comapring the predicted data with the test data
```

```python
y_test
```

```python
#Representing it in dataframe form by taking the test data(y_test) as actual and predicted profit (y_pred) as predicted profit

df1= pd.DataFrame({'Actual_Profit':y_test,'Predicted_profit':y_pred})
df1
```

```python
#now let us try to plot the predicted data
from IPython.core.pylabtools import figsize
df1.plot(figsize(20,30))
```

```python
#To check the score of determination or the accuracy of the predicted profit
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#Therefore the accuracy of the predicted profit is approx. 93.25%
```
